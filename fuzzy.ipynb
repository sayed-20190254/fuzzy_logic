```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

```python
Cholesterol = ctrl.Antecedent(np.arange(100, 500, 10), 'Cholesterol')
BloodPressure = ctrl.Antecedent(np.arange(115, 200, 5), 'BloodPressure')
PhysicalActivity = ctrl.Antecedent(np.arange(0, 1, .1), 'PhysicalActivity')
Age = ctrl.Antecedent(np.arange(0, 80,2), 'Age')
Smoking = ctrl.Antecedent(np.arange(0, 1, .1), 'Smoking')
Diabetes = ctrl.Antecedent(np.arange(70, 400, 10), 'Diabetes')
presenceOfHeart=ctrl.Consequent(np.arange(0,4,0.1),'presenceOfHeart')

presenceOfHeart['Healthy'] = fuzz.trapmf(presenceOfHeart.universe, [0,0,.5, 1.75])
presenceOfHeart['EarlyStage'] = fuzz.trimf(presenceOfHeart.universe, [1.5,2, 2.5])
presenceOfHeart['Advaned_stage'] = fuzz.trapmf(presenceOfHeart.universe, [2.3,3, 4,4])
presenceOfHeart.view()

Diabetes['Normal'] = fuzz.trimf(Diabetes.universe, [70,70, 120])
Diabetes['medium'] = fuzz.trimf(Diabetes.universe, [120, 150, 180])
Diabetes['High'] = fuzz.trimf(Diabetes.universe, [180, 290, 400])
Diabetes.view()

Cholesterol['Normal'] = fuzz.trimf(Cholesterol.universe, [100, 100, 200])
Cholesterol['medium'] = fuzz.trimf(Cholesterol.universe, [190, 220, 250])
Cholesterol['High'] = fuzz.trimf(Cholesterol.universe, [230, 270, 320])
Cholesterol['VeryHigh'] = fuzz.trimf(Cholesterol.universe, [280, 390, 500])
Cholesterol.view()

BloodPressure['Normal'] = fuzz.trimf(BloodPressure.universe, [115, 115, 130])
BloodPressure['medium'] = fuzz.trimf(BloodPressure.universe, [120, 135, 150])
BloodPressure['High'] = fuzz.trimf(BloodPressure.universe, [150, 175, 200])
BloodPressure.view()

PhysicalActivity['True'] = fuzz.trapmf(PhysicalActivity.universe, [0.3, 0.6, 1,1])

Age['Young'] = fuzz.trimf(Age.universe, [0, 0, 38])
Age['medium'] = fuzz.trimf(Age.universe, [34, 38, 45])
Age['Old'] = fuzz.trimf(Age.universe, [40, 60, 80])

Smoking['True'] = fuzz.trapmf(Smoking.universe, [0.3, 0.6, 1,1])

```

```python
Smoking.view()
BMI.view()
Age.view()
PhysicalActivity.view()
```

```python
rule1 = ctrl.Rule(Age['Old'] &BloodPressure['High']&Cholesterol['Normal']&Diabetes['Normal']&(~PhysicalActivity['True'])&(~Smoking['True']), presenceOfHeart['Healthy'])
rule2 = ctrl.Rule(Age['Young'] & BloodPressure['High']&Cholesterol['High']&Diabetes['High']&(~PhysicalActivity['True'])&(Smoking['True']), presenceOfHeart['EarlyStage'])
rule3 = ctrl.Rule(Age['medium'] & BloodPressure['High']&Cholesterol['High']&Diabetes['medium']&(~PhysicalActivity['True'])&(Smoking['True']), presenceOfHeart['EarlyStage'])
rule4 = ctrl.Rule(Age['Old'] & BloodPressure['High']&Cholesterol['High']&Diabetes['High']&(~PhysicalActivity['True'])&(Smoking['True']), presenceOfHeart['EarlyStage'])
rule5 = ctrl.Rule(Age['Old'] & BloodPressure['High']&Cholesterol['High']&Diabetes['Normal']&(~PhysicalActivity['True'])&(Smoking['True']), presenceOfHeart['EarlyStage'])
rule6 = ctrl.Rule(Age['Old'] & BloodPressure['Normal']&Cholesterol['VeryHigh']&Diabetes['Normal']&(PhysicalActivity['True']), presenceOfHeart['EarlyStage'])
rule7 = ctrl.Rule(Age['Old'] & BloodPressure['medium']&Cholesterol['VeryHigh']&Diabetes['medium']&(PhysicalActivity['True']), presenceOfHeart['EarlyStage'])
rule8 = ctrl.Rule(Age['Old'] & BloodPressure['High']&Cholesterol['VeryHigh']&Diabetes['Normal']&(~PhysicalActivity['True'])&(Smoking['True']), presenceOfHeart['Advaned_stage'])
rule9 = ctrl.Rule(Age['Old'] & BloodPressure['High']&(Smoking['True']), presenceOfHeart['EarlyStage'])
rule10 = ctrl.Rule(Age['Old'] & BloodPressure['High']&Diabetes['High'], presenceOfHeart['Advaned_stage'])
rule11 = ctrl.Rule(Age['Young'] & BloodPressure['Normal']&Diabetes['Normal']&(~Smoking['True']), presenceOfHeart['Healthy'])
rule12 = ctrl.Rule(Age['Young'] & BloodPressure['Normal']&Cholesterol['Normal']&(PhysicalActivity['True'])&(~Smoking['True']), presenceOfHeart['Healthy'])
rule13 = ctrl.Rule(Age['medium'] & BloodPressure['Normal']&Cholesterol['Normal']&Diabetes['Normal']&(PhysicalActivity['True'])&(~Smoking['True']), presenceOfHeart['Healthy'])
rule14 = ctrl.Rule(Age['Old'] & BloodPressure['Normal']&Cholesterol['Normal']&Diabetes['Normal']&(PhysicalActivity['True'])&(~Smoking['True']), presenceOfHeart['Healthy'])
```

```python
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['Age'] = 60
tipping.input['BloodPressure'] = 190
tipping.input['Cholesterol'] = 400
tipping.input['Diabetes'] = 300
tipping.input['PhysicalActivity'] = .5
tipping.input['Smoking'] = .75


tipping.compute()
print (tipping.output['presenceOfHeart'])

presenceOfHeart.view(sim=tipping)
```
